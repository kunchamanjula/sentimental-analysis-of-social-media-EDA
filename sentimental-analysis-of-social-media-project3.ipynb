# Sentimental analysis :

* Sentiment analysis is also known as opinion mining or emotion artificial intelligence.

 2.print the ratios of each user_friendly platform in social media platforms of dataset?

```python
def user_friendly_platform(a):
    Twitter = Instagram = Facebook = 0
    l=len(a)
    for i in a["Platforms"]:
        if i==' Twitter ':
            Twitter=Twitter+1 
        elif i==' Instagram ':
            Instagram=Instagram+1
        else:
            Facebook=Facebook+1
    print(Twitter/l)
    print(Instagram/l)
    print(Facebook/l)
print(user_friendly_platform(a))#ratios of platforms used by users based on social media platform
```

9. plot pie chart based on platforms used in social media analysis?

```python
Instagram=a.loc[a["Platforms"]==' Instagram '].count()[0] #splitting instagram from data
Facebook=a.loc[a["Platforms"]==' Facebook '].count()[0]#splitting facebook from data
Twitter=a.loc[a["Platforms"]==' Twitter '].count()[0]#splitting twitter from data
labels=["Instagram","Facebook","Twitter"] #giving labels paramerts in graph
plt.pie([Instagram,Facebook,Twitter],labels=labels,autopct="%2f%%")#plotting pie chart by passing arguments
plt.title("Platforms in social media analysis",fontdict={"fontsize":20,"fontname":"comic sans MS"}) #giving title to the graph
plt.style.use('ggplot')
plt.figure(figsize=(5,8),dpi=300) #most common dpi is 300
plt.show()
```
